```python
import pandas as pd
import pickle

f_location = "https://drive.google.com/file/d/1abp6Oc3XLvN7x2l2r1RHMjWNwoCTanLV/view?usp=sharing"
data_raw = pd.read_csv('https://drive.google.com/uc?export=download&id=' + f_location.split('/')[-2]) #Convert to pandas DataFrames

# Load debugging dataset (random sample of 2000 rows)
debugging_dataset = data_raw.sample(n=2000)

# Pickle the data for debugging dataset
debugging_dataset.to_pickle('debugging_dataset.pkl')

# Save working dataset
data_raw.to_pickle('working_dataset.pkl')

print("Datasets have been processed and pickled.")
```
